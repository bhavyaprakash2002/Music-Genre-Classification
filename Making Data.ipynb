### mfccs, chromagram, mel_spectrogram, tempo, zero_crossings

```python
mfccs = librosa.feature.mfcc(y=y, sr=sr)
chromagram = librosa.feature.chroma_stft(y=y, sr=sr)
mel_spectrogram = librosa.feature.melspectrogram(y=y, sr=sr)
# mfccs : Mel-Frequency Cepstral Coefficients = It is the short-term power spectrum of the sound signal
#     The process involves taking the Fourier transform of the signal, mapping it onto the mel scale, 
#     and then taking the logarithm of the powers at each mel frequency bin.

# chromagram : A chromagram is a representation of the 12 different pitch classes in an audio signal.
# mel_spectrogram : a mel spectrogram is a visual representation of the spectrum of a signal over time.
# tempo : Tempo refers to the speed or pace of a piece of music, typically measured in beats per minute (BPM).
# zero-crossings : Zero crossings are points in an audio signal where the amplitude changes its sign (crosses zero).
# rms : The energy of the audio signal.
 
```

```python
print(mfccs.shape)
print(chromagram.shape)
print(mel_spectrogram.shape)
```
