### Neural Networks

```python
import tensorflow as tf
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
y_train_mfccs = tf.keras.utils.to_categorical(y_train)
y_test_mfccs = tf.keras.utils.to_categorical(y_test)
```

```python
model = Sequential()
```

```python
model.add(Dense(15, input_shape = (20,), activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(6, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train_mfccs, epochs = 100, batch_size = 1024)
```

```python
model.evaluate(x_test, y_test_mfccs)
```
