```python
# import numpy as np
# import matplotlib.pyplot as plt
# import librosa.display

# Load the audio file
audio_file = r"C:\Users\Bhavya Prakash\Downloads\Folder 1\Folder 1\Funk.mp3"
y, sr = librosa.load(audio_file)

# Compute the spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# Plot the spectrogram
plt.figure(figsize=(20, 6))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')

# Add contour lines to the plot
plt.contour(D, colors='white', linewidths=1, alpha=0.1, extent=[0, librosa.get_duration(y=y, sr=sr), 0, 11025])

# Customize the plot
plt.title('Spectrogram with Contour Lines')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')

# Show the plot
plt.show()
```

```python
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
plt.figure(figsize=(20, 6))
librosa.display.specshow(mel_spec_db, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.show()
```

```python
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
plt.figure(figsize=(20, 6))
librosa.display.specshow(chroma, sr=sr, x_axis='time', y_axis='chroma')
plt.colorbar()
plt.title('Chromagram')
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Load the audio file
audio_file = r"C:\Users\Bhavya Prakash\Downloads\Folder 1\Folder 1\Funk.mp3"
y, sr = librosa.load(audio_file)

# Compute the spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

# Create a meshgrid for time and frequency
time = np.arange(0, D.shape[1])
frequency = np.arange(0, D.shape[0])
time, frequency = np.meshgrid(time, frequency)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(time, frequency, D, cmap='viridis', edgecolor='k')

# Customize the plot
ax.set_xlabel('Time (s)')
ax.set_ylabel('Frequency (Hz)')
ax.set_zlabel('Amplitude (dB)')
ax.set_title('3D Spectrogram')

# Show the plot
plt.show()
```
